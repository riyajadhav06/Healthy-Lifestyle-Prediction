# Preprocessing 

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
imputer=SimpleImputer(strategy="mean")
scaler=StandardScaler()
encoder=OneHotEncoder(handle_unknown="ignore",sparse=False)
```

```python
parameters = {"penalty ": ["11","12",'elasticnet'], "C" : [1,2,3,4,5,6,10,20,30,40,50], "max_iter": [100,200,300]}
```

```python
from sklearn import linear_model
from sklearn.model_selection import GridSearchCV
logistic_model =  linear_model.LogisticRegression()
```

## Create a model using Decision Tree

```python
parameter = {'criterion': ['gini','entropy','log_loss'],
            'splitter': ['best','random'],
            'max_depth':[1,3,5,7,9],
            'max_features':['auto','sqrt','log2']}
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
decision_tree = DecisionTreeClassifier()
```

```python
decision_tree = GridSearchCV(decision_tree,param_grid = parameter , cv = 3 , scoring = 'accuracy',verbose = 5)
decision_tree.fit(X_train_input, y_train)
```

```python
decision_tree.best_params_
```

```python
decision_tree = DecisionTreeClassifier(criterion = 'gini',max_depth = 9 , max_features = 'auto',
                                       splitter = 'best',random_state = 42)
decision_tree.fit(X_train_input, y_train)
```

```python
y_train_pred2 = decision_tree.predict(X_train_input)
```

```python
from sklearn.metrics import confusion_matrix
  
cm = confusion_matrix(y_train, y_train_pred2)
print ("Confusion Matrix : \n", cm)
```

```python
from sklearn.metrics import classification_report , accuracy_score , precision_score , recall_score , f1_score
print(accuracy_score(y_train_pred2,y_train))
print(precision_score(y_train_pred2,y_train))
print(recall_score(y_train_pred2,y_train))
print(f1_score(y_train_pred2,y_train))
```

```python
y_test_pred2 = decision_tree.predict(X_test_input)
```

```python
print(classification_report( y_train_pred2, y_train))
```
